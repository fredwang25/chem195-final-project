```python
# SIMILARITY

a_c = []
for i in range(100):
    a1 = np.random.binomial(n=1, p=0.5, size=(4,4)) * 255
    img = Image.fromarray(np.uint8(a1), 'L')
    #img.show()

    a2 = np.random.binomial(n=1, p=0.5, size=(4,4)) * 255
    img = Image.fromarray(np.uint8(a2), 'L')
    #img.show()

    a1 = a1.astype("float64").flatten()
    norm = np.sqrt(np.sum(a1 ** 2))
    a1 /= norm
    print(np.linalg.norm(a1))

    a2 = a2.astype("float64").flatten()
    norm = np.sqrt(np.sum(a2 ** 2))
    a2 /= norm
    print(np.linalg.norm(a2))

    from qiskit import transpile
    from qiskit.providers.aer import QasmSimulator
    from qiskit import QuantumRegister, ClassicalRegister
    from qiskit import QuantumCircuit

    num_qubits = 4

    simulator = QasmSimulator()
    qc = QuantumCircuit(QuantumRegister(num_qubits,'q'))
    qc.initialize(a1.flatten().tolist())

    qc2 = QuantumCircuit(QuantumRegister(num_qubits,'q'))
    qc2.initialize(a2.flatten().tolist())

    swap_test_circuit = QuantumCircuit(2*num_qubits + 1, 1)
    swap_test_circuit.compose(qc, qubits=list(range(1,num_qubits+1)), inplace=True )
    swap_test_circuit.compose(qc2, qubits=list(range(num_qubits+1, 2*num_qubits+1)), inplace=True)
    swap_test_circuit.h(0)
    for i in range(num_qubits):
        swap_test_circuit.cswap(0,i+1,i+1+num_qubits)
    swap_test_circuit.h(0)
    swap_test_circuit.measure([0],[0])
    # qc.measure_all()


    # # # compile the circuit down to low-level QASM instructions
    # # # supported by the backend (not needed for simple circuits)
    compiled_circuit = transpile(swap_test_circuit, simulator)

    # # Execute the circuit on the qasm simulator
    job = simulator.run(compiled_circuit, shots=100)

    # # Grab results from the job
    result = job.result()

    # # Returns counts
    # print(result)
    counts = result.get_counts(compiled_circuit)
    print("\nCounts:",counts)
    a_c.append(counts['1'])
```

```python
a_c=np.array(a_c)
print(np.mean(a_c))
print(np.std(a_c))
```

```python
# SYMMETRY

a_c = []
for i in range(100):
    a1 = np.random.binomial(n=1, p=0.5, size=(4,4)) * 255
    img = Image.fromarray(np.uint8(a1), 'L')
    #img.show()

    a2 = np.random.binomial(n=1, p=0.5, size=(4,4)) * 255
    img = Image.fromarray(np.uint8(a2), 'L')
    #img.show()

    a1 = a1.astype("float64").flatten()
    norm = np.sqrt(np.sum(a1 ** 2))
    a1 /= norm
    print(np.linalg.norm(a1))

    a2 = a2.astype("float64").flatten()
    norm = np.sqrt(np.sum(a2 ** 2))
    a2 /= norm
    print(np.linalg.norm(a2))

    from qiskit import transpile
    from qiskit.providers.aer import QasmSimulator
    from qiskit import QuantumRegister, ClassicalRegister
    from qiskit import QuantumCircuit

    num_qubits = 4

    simulator = QasmSimulator()
    qc = QuantumCircuit(QuantumRegister(num_qubits,'q'))
    qc.initialize(a1.flatten().tolist())
    for i in range(num_qubits):
        qc.x(i)

    qc2 = QuantumCircuit(QuantumRegister(num_qubits,'q'))
    qc2.initialize(a2.flatten().tolist())
    for i in range(num_qubits):
        qc2.id(i)

    swap_test_circuit = QuantumCircuit(2*num_qubits + 1, 1)
    swap_test_circuit.compose(qc, qubits=list(range(1,num_qubits+1)), inplace=True )
    swap_test_circuit.compose(qc2, qubits=list(range(num_qubits+1, 2*num_qubits+1)), inplace=True)
    swap_test_circuit.h(0)
    for i in range(num_qubits):
        swap_test_circuit.cswap(0,i+1,i+1+num_qubits)
    swap_test_circuit.h(0)
    swap_test_circuit.measure([0],[0])
    # qc.measure_all()


    # # # compile the circuit down to low-level QASM instructions
    # # # supported by the backend (not needed for simple circuits)
    compiled_circuit = transpile(swap_test_circuit, simulator)

    # # Execute the circuit on the qasm simulator
    job = simulator.run(compiled_circuit, shots=100)

    # # Grab results from the job
    result = job.result()

    # # Returns counts
    # print(result)
    counts = result.get_counts(compiled_circuit)
    print("\nCounts:",counts)
    a_c.append(counts['1'])
```

```python
a_c=np.array(a_c)
print(np.mean(a_c))
print(np.std(a_c))
```
